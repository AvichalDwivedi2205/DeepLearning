```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
# We are making the inputs between 0 and 1
```

```python
y_train[2]
```

```python
y_train[4]
```

```python
X_train_flattened = X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
print(X_train_flattened.shape)
print(X_test_flattened.shape)
# This just flattens the 28*28 into 784
```

```python
X_train[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train_flattened, y_train, epochs=5)
#Dense meand every neuron is connected to every other neuron in the next layer
#Sequential means We have a stack of layers in my neural network
#output shape is 10 that is 0 to 9. The 10 output neurons are from 0 to 9
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
model.predict(X_test_flattened)
```

```python
plt.matshow(X_test[1])
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
# To find the maximum value of probability in an array
# This finds the highest probability of the 10 neurons taht is from 0 to 10 to give the answer
np.argmax(y_predicted[1])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train_flattened, y_train, epochs=5)
#Here we will add hidden layer to this
```

```python
model.evaluate(X_test_flattened, y_test)
y_predicted = model.predict(X_test_flattened)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),# Here we can flatten an array using the keras flatte n 
    keras.layers.Dense(256, activation='tanh'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train, y_train, epochs=10)
#Here we will use keras flattened layer
```
